# 3. Definición de la variable objetivo

El objetivo predictivo de este trabajo es estimar la supervivencia a cinco años tras el diagnóstico de cáncer de mama.

Para ello, se construye una variable binaria que indica si la paciente ha sobrevivido al menos 60 meses desde el diagnóstico.

La definición de esta variable requiere especial cuidado, ya que en los datos clínicos pueden existir casos censurados. En particular, los pacientes que siguen vivos pero tienen menos de 60 meses de seguimiento no pueden clasificarse de forma fiable como supervivientes o no supervivientes a cinco años.

Por este motivo, se aplica el siguiente criterio:

- `1`: pacientes con supervivencia igual o superior a 60 meses.
- `0`: pacientes fallecidos antes de alcanzar los 60 meses.
- No clasificables: pacientes vivos con menos de 60 meses de seguimiento.

Los casos no clasificables se excluyen del modelado supervisado, ya que no permiten asignar una etiqueta objetivo fiable.

## 3.3 Construcción de la variable objetivo

## 3.8 Justificación de la formulación como clasificación binaria

La formulación de la supervivencia a cinco años como una tarea de clasificación binaria constituye una decisión metodológica consciente. Esta formulación permite aplicar un marco común a la regresión logística, Random Forest y XGBoost, así como comparar su rendimiento mediante métricas estándar de clasificación.

La pregunta respondida por el modelo es concreta: con la información clínico-epidemiológica disponible, ¿el paciente alcanza o no los 60 meses de supervivencia global tras el diagnóstico? Este planteamiento facilita el análisis del desbalance, la comparación de algoritmos, el ajuste del umbral de decisión, la calibración y el estudio de los errores.

La simplificación tiene, sin embargo, limitaciones importantes. Un análisis de supervivencia mediante Kaplan–Meier, regresión de Cox, Random Survival Forest, XGBoost con objetivos de supervivencia, DeepSurv o modelos de riesgos competitivos permitiría incorporar el tiempo hasta el evento y gestionar directamente la censura. Estos métodos responderían a una pregunta diferente y más completa: cómo evoluciona el riesgo a lo largo del tiempo.

Por tanto, la clasificación a horizonte fijo no se plantea como sustituto de un análisis de supervivencia ni como el enfoque clínico ideal, sino como una simplificación delimitada y operativa para el alcance de este TFM. La comparación con modelos específicos de supervivencia constituye una extensión natural del trabajo.

## 4.3 Funciones auxiliares para el análisis exploratorio

Para mantener el notebook ordenado y evitar repetir código en el análisis de variables categóricas, se definen funciones auxiliares.

Estas funciones permiten obtener, para cada variable, una tabla con el número de observaciones por categoría, el porcentaje que representa cada categoría dentro del dataset y la tasa media de supervivencia a cinco años.

Además, se define una función de visualización que representa la tasa de supervivencia a cinco años por categoría. Esta representación facilita la identificación de posibles diferencias entre grupos antes de la fase de modelado.

```python
def categorical_summary(df, column, target="survived_5_years"):
    """
    Devuelve una tabla resumen con:
    - número de observaciones por categoría
    - porcentaje sobre el total del dataset
    - tasa media de supervivencia a cinco años
    """
    summary = (
        df
        .groupby(column)
        .agg(
            count=(target, "size"),
            survival_5y_rate=(target, "mean")
        )
        .reset_index()
        .sort_values("count", ascending=False)
    )

    summary["percentage"] = (
        summary["count"] / len(df) * 100
    ).round(2)

    summary["survival_5y_rate"] = (
        summary["survival_5y_rate"] * 100
    ).round(2)

    return summary


def plot_survival_rate_by_category(df, column, top_n=15, target="survived_5_years"):
    """
    Representa la tasa de supervivencia a cinco años por categoría.

    Para variables con muchas categorías, permite mostrar únicamente las categorías
    más frecuentes mediante el parámetro top_n.
    """
    summary = categorical_summary(df, column, target=target)

    if top_n is not None:
        summary = summary.head(top_n)

    plt.figure(figsize=(10, 5))

    plt.bar(
        summary[column].astype(str),
        summary["survival_5y_rate"]
    )

    plt.title(f"Tasa de supervivencia a cinco años según {column}")
    plt.xlabel(column)
    plt.ylabel("Supervivencia a cinco años (%)")
    plt.xticks(rotation=45, ha="right")
    plt.ylim(0, 100)
    plt.tight_layout()
    plt.show()

    return summary
```

### 4.12.1 Calidad, disponibilidad y valores especiales por variable

La ausencia de información en SEER no se representa únicamente mediante valores `NaN`. Algunas variables contienen códigos explícitos de información desconocida o no evaluable, mientras que otras combinan en una misma categoría la ausencia de tratamiento y la falta de información.

Por este motivo se distinguen tres conceptos:

- Nulos explícitos: valores `NaN`.
- No informados: códigos que representan claramente información desconocida o no disponible.
- Categorías ambiguas: valores como `No/Unknown` o `None/Unknown`, que no permiten diferenciar entre ausencia de tratamiento e información desconocida.

Se utiliza un 75 % de información disponible como umbral orientativo para detectar variables problemáticas, pero no como una regla automática de eliminación. La utilidad predictiva, el significado clínico y el posible carácter informativo de la ausencia también deben considerarse.

```python
clear_missing_codes = {
    "race": {"Unknown"},
    "marital_status": {"Unknown"},
    "tumor_size": {"Blank(s)", "999"},
    "summary_stage": {"Blank(s)", "Unknown/unstaged"},
    "surgery_primary_site": {"99"}
}

ambiguous_codes = {
    "radiation": {
        "None/Unknown",
        "Recommended, unknown if administered"
    },
    "chemotherapy": {
        "No/Unknown"
    }
}

quality_rows = []

predictor_columns = [
    col for col in df_model_main.columns
    if col != "survived_5_years"
]

for col in predictor_columns:
    series = df_model_main[col]
    series_as_text = series.astype("string")

    n_null = int(series.isna().sum())

    n_special_missing = int(
        series_as_text.isin(clear_missing_codes.get(col, set())).sum()
    )

    n_ambiguous = int(
        series_as_text.isin(ambiguous_codes.get(col, set())).sum()
    )

    n_not_available = n_null + n_special_missing
    pct_not_available = n_not_available / len(df_model_main) * 100
    pct_available = 100 - pct_not_available
    pct_ambiguous = n_ambiguous / len(df_model_main) * 100

    if pct_available < 75:
        decision = "Alerta <75 %; evaluar exclusión en análisis de sensibilidad"
    elif n_ambiguous > 0:
        decision = "Mantener como categoría explícita; interpretación cautelosa"
    else:
        decision = "Mantener"

    quality_rows.append({
        "variable": col,
        "n_categorias": series.nunique(dropna=False),
        "n_nulos_explicitos": n_null,
        "n_especiales_no_informados": n_special_missing,
        "pct_no_disponible": pct_not_available,
        "pct_disponible": pct_available,
        "n_categorias_ambiguas": n_ambiguous,
        "pct_categoria_ambigua": pct_ambiguous,
        "decision": decision
    })

quality_final = (
    pd.DataFrame(quality_rows)
    .sort_values(
        ["pct_disponible", "n_categorias"],
        ascending=[True, False]
    )
    .reset_index(drop=True)
)

quality_final[
    [
        "pct_no_disponible",
        "pct_disponible",
        "pct_categoria_ambigua"
    ]
] = quality_final[
    [
        "pct_no_disponible",
        "pct_disponible",
        "pct_categoria_ambigua"
    ]
].round(2)

quality_final
```

La tabla permite diferenciar variables con ausencia inequívoca de aquellas cuyas categorías son ambiguas. Por ello se mantendrá en el análisis principal para conservar la comparabilidad inicial, pero se excluirá posteriormente en un análisis de sensibilidad.

### 4.12.2 Asociación entre predictores categóricos

La mayor parte de los predictores son categóricos, por lo que no resulta adecuado calcular una correlación de Pearson sobre sus códigos numéricos. Los códigos de localización, histología o cirugía no representan magnitudes continuas ni distancias numéricas interpretables.

Para estudiar la redundancia entre predictores se utiliza la V de Cramér corregida, una medida de asociación entre variables categóricas con valores comprendidos entre 0 y 1. Para reducir el coste computacional, la matriz se estima sobre una muestra aleatoria reproducible de hasta 100.000 observaciones.

```python
from scipy.stats import chi2_contingency
import os

os.makedirs("figuras", exist_ok=True)

def cramers_v_corrected(x, y):
    """V de Cramér con corrección del sesgo por tamaño y cardinalidad."""
    contingency = pd.crosstab(x, y, dropna=False)

    if contingency.shape[0] < 2 or contingency.shape[1] < 2:
        return np.nan

    chi2 = chi2_contingency(contingency, correction=False)[0]
    n = contingency.to_numpy().sum()
    r, k = contingency.shape

    if n <= 1:
        return np.nan

    phi2 = chi2 / n
    phi2_corrected = max(
        0,
        phi2 - ((k - 1) * (r - 1)) / (n - 1)
    )

    r_corrected = r - ((r - 1) ** 2) / (n - 1)
    k_corrected = k - ((k - 1) ** 2) / (n - 1)

    denominator = min(k_corrected - 1, r_corrected - 1)

    if denominator <= 0:
        return np.nan

    return np.sqrt(phi2_corrected / denominator)


cramer_features = [
    "age_group",
    "sex",
    "race",
    "marital_status",
    "primary_site",
    "histologic_type",
    "tumor_size",
    "summary_stage",
    "surgery_primary_site",
    "radiation",
    "chemotherapy"
]

sample_size = min(100_000, len(df_model_main))

cramer_sample = df_model_main[
    cramer_features
].sample(
    n=sample_size,
    random_state=42
)

cramer_matrix = pd.DataFrame(
    np.eye(len(cramer_features)),
    index=cramer_features,
    columns=cramer_features
)

for i, col_a in enumerate(cramer_features):
    for j in range(i + 1, len(cramer_features)):
        col_b = cramer_features[j]

        value = cramers_v_corrected(
            cramer_sample[col_a],
            cramer_sample[col_b]
        )

        cramer_matrix.loc[col_a, col_b] = value
        cramer_matrix.loc[col_b, col_a] = value

plt.figure(figsize=(12, 10))
image = plt.imshow(
    cramer_matrix,
    vmin=0,
    vmax=1,
    cmap="viridis"
)

plt.colorbar(image, label="V de Cramér corregida")
plt.xticks(
    range(len(cramer_features)),
    cramer_features,
    rotation=90
)
plt.yticks(
    range(len(cramer_features)),
    cramer_features
)

plt.title(
    "Asociación entre predictores categóricos "
    "(V de Cramér corregida)"
)
plt.tight_layout()
plt.savefig(
    "figuras/cramers_v_predictors.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()

high_associations = []

for i, col_a in enumerate(cramer_features):
    for j in range(i + 1, len(cramer_features)):
        col_b = cramer_features[j]
        value = cramer_matrix.loc[col_a, col_b]

        if pd.notna(value) and value >= 0.50:
            high_associations.append({
                "variable_1": col_a,
                "variable_2": col_b,
                "cramers_v": value
            })

high_associations_df = (
    pd.DataFrame(high_associations)
    .sort_values(
        "cramers_v",
        ascending=False
    )
    .reset_index(drop=True)
    if high_associations
    else pd.DataFrame(
        columns=["variable_1", "variable_2", "cramers_v"]
    )
)

high_associations_df.round(4)
```

La V de Cramér permite detectar asociaciones y posibles redundancias entre predictores categóricos. Una asociación elevada no implica que una de las variables deba eliminarse automáticamente: antes de hacerlo deben considerarse su significado clínico, su disponibilidad y si ambas representan realmente el mismo concepto.

La redundancia puede afectar especialmente a la estabilidad de los coeficientes de la regresión logística. En modelos basados en árboles no implica necesariamente sobreajuste, aunque puede repartir la importancia entre variables relacionadas y dificultar su interpretación. Por ello, los resultados se utilizan como diagnóstico metodológico y no como una regla automática de selección.

### 4.12.3 Tests de asociación chi-cuadrado

Para cada variable categórica se construye una tabla de contingencia frente a `survived_5_years` y se aplica la prueba chi-cuadrado de independencia.

Dado el enorme tamaño muestral de la cohorte, los p-valores pueden resultar extremadamente pequeños incluso para asociaciones de magnitud reducida. Por ello, la interpretación no se basa únicamente en la significación estadística, sino también en la V de Cramér corregida y en las diferencias observadas entre categorías.

Se informa además de la frecuencia esperada mínima. Cuando esta es inferior a 5, la aproximación asintótica del chi-cuadrado puede ser menos fiable para las categorías poco frecuentes, por lo que conviene considerar su agrupación.

```python
cat_vars = [
    "age_group",
    "sex",
    "race",
    "marital_status",
    "primary_site",
    "histologic_type",
    "tumor_size",
    "summary_stage",
    "surgery_primary_site",
    "radiation",
    "chemotherapy"
]

assoc_rows = []

for col in cat_vars:
    contingency = pd.crosstab(
        df_model_main[col],
        df_model_main["survived_5_years"]
    )

    chi2, p_value, dof, expected = chi2_contingency(contingency)
    min_expected = expected.min()

    assoc_rows.append({
        "variable": col,
        "n_categorias": int(contingency.shape[0]),
        "chi2": chi2,
        "grados_libertad": int(dof),
        "p_value": p_value,
        "min_frecuencia_esperada": min_expected,
        "V_Cramer_corregida": cramers_v_corrected(
            df_model_main[col],
            df_model_main["survived_5_years"]
        ),
        "advertencia": (
            "Revisar categorías infrecuentes"
            if min_expected < 5
            else "Sin advertencia"
        )
    })

assoc_df = (
    pd.DataFrame(assoc_rows)
    .sort_values(
        "V_Cramer_corregida",
        ascending=False
    )
    .reset_index(drop=True)
)

assoc_df[
    [
        "chi2",
        "min_frecuencia_esperada",
        "V_Cramer_corregida"
    ]
] = assoc_df[
    [
        "chi2",
        "min_frecuencia_esperada",
        "V_Cramer_corregida"
    ]
].round(4)

assoc_df
```

### 4.12.4 Odds ratios y forest plot

Para las variables clínicas de menor cardinalidad se calculan *odds ratios* (OR) del evento **no supervivencia a cinco años** (clase 0) de cada categoría frente a una categoría de referencia (la más frecuente de cada variable; para la edad se toma `50-54 years`). Los intervalos de confianza al 95 % se obtienen por el método de Woolf sobre tablas 2×2, con corrección de Haldane-Anscombe cuando alguna casilla es cero. Un OR mayor que 1 indica mayor riesgo de no supervivencia respecto a la referencia; un OR menor que 1, menor riesgo.

Las variables de tratamiento proceden de un registro observacional, por lo que estos OR describen **asociaciones**, no efectos causales.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
os.makedirs("figuras", exist_ok=True)

def odds_ratios_vs_ref(data, col, target="survived_5_years", ref_level=None, min_count=300):
    """OR de no supervivencia (clase 0) de cada nivel frente a la referencia.
    IC95% por el metodo de Woolf; correccion de Haldane-Anscombe si hay casillas a 0.
    La referencia es el nivel mas frecuente salvo que se indique."""
    vc = data[col].value_counts()
    levels = vc[vc >= min_count].index.tolist()
    if ref_level is None:
        ref_level = vc.index[0]
    ref = data[data[col] == ref_level]
    a_ref = (ref[target] == 0).sum(); b_ref = (ref[target] == 1).sum()
    out = []
    for lev in levels:
        if lev == ref_level:
            continue
        sub = data[data[col] == lev]
        a = (sub[target] == 0).sum(); b = (sub[target] == 1).sum()
        t = np.array([[a, b], [a_ref, b_ref]], dtype=float)
        if (t == 0).any():
            t += 0.5
        orr = (t[0, 0] * t[1, 1]) / (t[0, 1] * t[1, 0])
        se = np.sqrt((1.0 / t).sum())
        lo, hi = np.exp(np.log(orr) - 1.96 * se), np.exp(np.log(orr) + 1.96 * se)
        out.append({"variable": col, "nivel": str(lev), "referencia": str(ref_level),
                    "OR": round(orr, 3), "IC95_inf": round(lo, 3),
                    "IC95_sup": round(hi, 3), "n": int(a + b)})
    return pd.DataFrame(out)

forest_df = pd.concat([
    odds_ratios_vs_ref(df_model_main, "summary_stage"),
    odds_ratios_vs_ref(df_model_main, "age_group", ref_level="50-54 years"),
    odds_ratios_vs_ref(df_model_main, "sex"),
    odds_ratios_vs_ref(df_model_main, "race"),
    odds_ratios_vs_ref(df_model_main, "marital_status"),
    odds_ratios_vs_ref(df_model_main, "radiation"),
    odds_ratios_vs_ref(df_model_main, "chemotherapy"),
], ignore_index=True)

orden = ["summary_stage", "age_group", "sex", "race", "marital_status", "radiation", "chemotherapy"]
forest_df["__v"] = pd.Categorical(forest_df["variable"], categories=orden, ordered=True)
forest_df = forest_df.sort_values(["__v", "OR"]).drop(columns="__v").reset_index(drop=True)

plt.figure(figsize=(9.5, max(5, 0.40 * len(forest_df))))
labels = forest_df["variable"] + " = " + forest_df["nivel"]
colors = ["#c0392b" if orr > 1 else "#2471a3" for orr in forest_df["OR"]]
for i, (orr, lo, hi, c) in enumerate(zip(forest_df["OR"], forest_df["IC95_inf"],
                                         forest_df["IC95_sup"], colors)):
    plt.plot([lo, hi], [i, i], color=c, lw=1.3, zorder=1)
    plt.scatter([orr], [i], color=c, s=28, zorder=2)
plt.axvline(1.0, linestyle="--", color="gray", lw=1)
plt.xscale("log")
plt.yticks(range(len(forest_df)), labels, fontsize=8)
plt.gca().invert_yaxis()
plt.xlabel("Odds ratio de NO supervivencia a 5 años (IC95%, escala log)")
plt.title("Asociación univariante de variables clínicas con la no supervivencia a 5 años")
plt.tight_layout()
plt.savefig("figuras/forest_plot_or.png", dpi=300, bbox_inches="tight")
plt.show()

forest_df
```

## 5.5 Pipeline de preprocesamiento

Se construye un pipeline reproducible mediante `ColumnTransformer`, de modo que todas las transformaciones se ajustan exclusivamente sobre el conjunto de entrenamiento.

Las variables categóricas se imputan mediante la categoría más frecuente únicamente como mecanismo defensivo frente a posibles valores `NaN`. La mayoría de ausencias en SEER están ya representadas por categorías explícitas, por lo que no son reemplazadas automáticamente.

La única variable numérica es `year_diagnosis`, que no presenta valores faltantes en la cohorte utilizada. La imputación mediante mediana se conserva como protección técnica del pipeline, pero no modifica ninguna observación en el análisis actual.

No se emplea imputación KNN porque:

- No existen valores faltantes efectivos en la única variable numérica.
- La mayoría de los predictores son categóricos.
- Los códigos clínicos no representan distancias numéricas adecuadas para un algoritmo de vecinos.
- Su aplicación sobre cerca de un millón de registros supondría un coste computacional elevado sin aportar una mejora justificada.

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

preprocessor = ColumnTransformer(
    transformers=[
        ("categorical", categorical_transformer, categorical_features),
        ("numeric", numeric_transformer, numeric_features)
    ]
)

preprocessor
```

El `ColumnTransformer` aplica transformaciones diferenciadas según el tipo de predictor. Las variables categóricas se codifican mediante One-Hot Encoding, mientras que `year_diagnosis` se estandariza.

`handle_unknown="ignore"` permite procesar categorías del conjunto de prueba que no estuvieran presentes durante el ajuste. El pipeline principal conserva todas las categorías observadas para mantener la definición original del experimento.

Posteriormente se realizará un análisis de sensibilidad alternativo en el que se excluirá `tumor_size`, se agrupará el estado civil y se combinarán automáticamente las categorías infrecuentes. Esta comparación permitirá comprobar si la alta dimensionalidad o la ausencia de información condicionan de forma relevante el rendimiento.

## 7.1 Función común de evaluación

Para comparar los modelos de forma homogénea, se define una función que calcula las principales métricas de clasificación sobre el conjunto de prueba.

La función devuelve:

- Accuracy.
- Precision.
- Recall.
- F1-score.
- ROC-AUC.
- Classification report.
- Matriz de confusión.

Esta función permitirá evaluar todos los modelos bajo el mismo criterio.

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    classification_report,
    confusion_matrix
)

def evaluate_model(model, X_test, y_test, model_name):
    """
    Evalúa un modelo de clasificación binaria sobre el conjunto de prueba.

    Parámetros:
    - model: pipeline o modelo entrenado.
    - X_test: variables predictoras del conjunto de prueba.
    - y_test: variable objetivo real.
    - model_name: nombre del modelo evaluado.

    Devuelve:
    - Diccionario con métricas principales.
    - Predicciones de clase.
    - Probabilidades estimadas para la clase positiva.
    - Matriz de confusión.
    """

    y_pred = model.predict(X_test)

    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)[:, 1]
    else:
        y_proba = model.decision_function(X_test)

    metrics = {
        "model": model_name,
        "accuracy": accuracy_score(y_test, y_pred),
        "precision": precision_score(y_test, y_pred),
        "recall": recall_score(y_test, y_pred),
        "f1_score": f1_score(y_test, y_pred),
        "roc_auc": roc_auc_score(y_test, y_proba)
    }

    cm = confusion_matrix(y_test, y_pred)

    print(f"===== {model_name} =====")
    print("\nClassification report:")
    print(classification_report(
        y_test,
        y_pred,
        target_names=["No sobrevive 5 años", "Sobrevive 5 años"]
    ))

    print("\nMatriz de confusión:")
    print(cm)

    return metrics, y_pred, y_proba, cm
```

La función anterior permite estandarizar la evaluación de los modelos. A partir de este punto, todos los algoritmos entrenados se evaluarán de la misma forma, facilitando la comparación posterior.

Además, las métricas se almacenarán en una lista para construir una tabla final comparativa.

## 7.6 Criterio de selección de métricas en contexto clínico

En un problema médico de predicción pronóstica, no todas las métricas tienen la misma importancia. La selección del modelo no debe basarse únicamente en la exactitud global, ya que el dataset presenta un desbalance moderado y la clase de mayor interés clínico es la de pacientes que no alcanzan la supervivencia a cinco años.

En este trabajo, la clase `0` representa a los pacientes que fallecen antes de los cinco años, mientras que la clase `1` representa a los pacientes que sobreviven al menos cinco años.

Desde una perspectiva clínica, uno de los errores más relevantes consiste en clasificar como superviviente a un paciente que realmente no alcanza la supervivencia a cinco años. Este tipo de error implica no identificar adecuadamente a un paciente de peor pronóstico.

Por este motivo, se presta especial atención a las métricas de la clase `No sobrevive 5 años`, especialmente:

- `recall` de la clase 0, porque mide la capacidad del modelo para identificar correctamente a los pacientes que realmente no sobreviven cinco años.
- `F1-score` de la clase 0, porque equilibra la detección de pacientes de alto riesgo con la precisión de esas predicciones.
- AUC-ROC, porque resume la capacidad discriminativa global del modelo a distintos umbrales de decisión.

La accuracy se mantiene como métrica descriptiva global, pero no se utiliza como criterio principal de selección del modelo, ya que puede verse favorecida por la clase mayoritaria.

```python
from sklearn.metrics import precision_recall_fscore_support

def build_clinical_metrics_row(
    model_name,
    y_true,
    y_pred,
    y_proba
):
    precision, recall, f1, _ = precision_recall_fscore_support(
        y_true,
        y_pred,
        labels=[0, 1],
        zero_division=0
    )

    return {
        "model": model_name,
        "precision_class_0": precision[0],
        "recall_class_0": recall[0],
        "f1_class_0": f1[0],
        "roc_auc": roc_auc_score(y_true, y_proba),
        "accuracy": accuracy_score(y_true, y_pred)
    }


clinical_metrics = pd.DataFrame([
    build_clinical_metrics_row(
        "Logistic Regression Baseline",
        y_test,
        y_pred_baseline,
        y_proba_baseline
    ),
    build_clinical_metrics_row(
        "Random Forest",
        y_test,
        y_pred_rf,
        y_proba_rf
    ),
    build_clinical_metrics_row(
        "XGBoost",
        y_test,
        y_pred_xgb,
        y_proba_xgb
    ),
    build_clinical_metrics_row(
        "XGBoost Optimizado",
        y_test,
        y_pred_best_xgb,
        y_proba_best_xgb
    )
]).round(4)

clinical_metrics.sort_values(
    by=["recall_class_0", "f1_class_0", "roc_auc"],
    ascending=False
)
```

Desde este criterio clínico, el XGBoost optimizado se posiciona como el modelo más interesante, ya que obtiene el mayor AUC-ROC y el mayor recall para la clase `No sobrevive 5 años`. Aunque su accuracy es inferior a la regresión logística, esta métrica no se considera prioritaria en este contexto debido al desbalance entre clases y a la relevancia clínica de identificar correctamente a los pacientes de peor pronóstico.

No obstante, la mejora respecto al baseline es moderada. La regresión logística sigue siendo un modelo de referencia muy competitivo por su simplicidad, estabilidad e interpretabilidad. Por tanto, la elección final no debe formularse únicamente como una comparación numérica, sino como un equilibrio entre rendimiento predictivo, interpretabilidad y utilidad clínica.

En este sentido, se consideran dos modelos especialmente relevantes para las siguientes fases:

- Regresión logística, como modelo simple, interpretable y con rendimiento competitivo.
- XGBoost optimizado, como modelo con mayor capacidad discriminativa global y mejor recall en la clase de mayor interés clínico.

Random Forest no mejora al baseline ni a XGBoost con la configuración utilizada. En conjunto, los resultados indican que las variables disponibles contienen señal predictiva, pero que la complejidad adicional de los modelos de árboles solo produce una mejora limitada.

```python
def bootstrap_auc_comparison(
    y_true,
    score_logistic,
    score_xgboost,
    n_boot=1000,
    seed=42
):
    """
    Estima mediante bootstrap:
    - IC95 % del AUC de la regresión logística.
    - IC95 % del AUC del XGBoost optimizado.
    - IC95 % pareado de la diferencia AUC_XGBoost - AUC_Logística.

    El remuestreo es pareado: en cada iteración se utilizan los mismos índices
    para ambos modelos, respetando la dependencia entre sus predicciones.
    """
    rng = np.random.RandomState(seed)
    y_true = np.asarray(y_true)
    score_logistic = np.asarray(score_logistic)
    score_xgboost = np.asarray(score_xgboost)

    auc_logistic_boot = []
    auc_xgboost_boot = []
    delta_boot = []

    for _ in range(n_boot):
        idx = rng.randint(0, len(y_true), len(y_true))
        y_sample = y_true[idx]

        # En muestras con una sola clase el AUC no está definido.
        if np.unique(y_sample).size < 2:
            continue

        auc_log = roc_auc_score(y_sample, score_logistic[idx])
        auc_xgb = roc_auc_score(y_sample, score_xgboost[idx])

        auc_logistic_boot.append(auc_log)
        auc_xgboost_boot.append(auc_xgb)
        delta_boot.append(auc_xgb - auc_log)

    auc_logistic = roc_auc_score(y_true, score_logistic)
    auc_xgboost = roc_auc_score(y_true, score_xgboost)
    delta_auc = auc_xgboost - auc_logistic

    ci_logistic = np.percentile(auc_logistic_boot, [2.5, 97.5])
    ci_xgboost = np.percentile(auc_xgboost_boot, [2.5, 97.5])
    ci_delta = np.percentile(delta_boot, [2.5, 97.5])

    summary = pd.DataFrame([
        {
            "estimacion": "AUC regresión logística",
            "valor": auc_logistic,
            "IC95_inf": ci_logistic[0],
            "IC95_sup": ci_logistic[1]
        },
        {
            "estimacion": "AUC XGBoost optimizado",
            "valor": auc_xgboost,
            "IC95_inf": ci_xgboost[0],
            "IC95_sup": ci_xgboost[1]
        },
        {
            "estimacion": "Diferencia pareada de AUC (XGBoost - logística)",
            "valor": delta_auc,
            "IC95_inf": ci_delta[0],
            "IC95_sup": ci_delta[1]
        }
    ]).round(4)

    display(summary)

    if ci_delta[0] > 0:
        print(
            "El IC95 % pareado de la diferencia no incluye 0: "
            "la ventaja discriminativa del XGBoost es compatible con una diferencia positiva."
        )
    elif ci_delta[1] < 0:
        print(
            "El IC95 % pareado de la diferencia no incluye 0 y favorece a la regresión logística."
        )
    else:
        print(
            "El IC95 % pareado de la diferencia incluye 0: "
            "no se demuestra una diferencia estadística entre ambos AUC."
        )

    print(
        "La magnitud observada debe valorarse también desde el punto de vista práctico, "
        "no solo estadístico."
    )

    return summary

bootstrap_auc_summary = bootstrap_auc_comparison(
    y_true=y_test,
    score_logistic=y_proba_baseline,
    score_xgboost=y_proba_best_xgb,
    n_boot=1000,
    seed=42
)
```

Para cuantificar la incertidumbre del rendimiento se estimaron intervalos de confianza del 95 % mediante 1.000 remuestreos bootstrap sobre el conjunto de prueba.

La regresión logística obtuvo un AUC-ROC de 0,8274, con un IC95 % de 0,8251–0,8297. El XGBoost optimizado alcanzó un AUC-ROC de 0,8316, con un IC95 % de 0,8294–0,8339.

La diferencia pareada fue de 0,0042 puntos de AUC a favor del XGBoost, con un IC95 % de 0,0037–0,0047. Este intervalo no contiene el valor cero, por lo que la diferencia discriminativa observada es compatible con una ventaja estadísticamente positiva del XGBoost.

No obstante, su magnitud es muy reducida. El gran tamaño muestral permite detectar diferencias pequeñas que no necesariamente poseen relevancia práctica. Por ello, la regresión logística continúa siendo una alternativa competitiva cuando se priorizan la sencillez y la interpretabilidad.

## 7.8 Ajuste del umbral de decisión

Hasta ahora, los modelos se han evaluado utilizando el umbral de decisión por defecto. En clasificación binaria, este umbral suele ser 0,5: si la probabilidad estimada de pertenecer a la clase positiva es igual o superior a 0,5, el modelo predice la clase `1`; en caso contrario, predice la clase `0`.

Sin embargo, en un problema médico este umbral no tiene por qué ser el más adecuado. En este trabajo, la clase `0` representa a los pacientes que no sobreviven cinco años, que es la clase de mayor interés clínico. Por tanto, puede ser conveniente ajustar el umbral para mejorar la detección de pacientes de peor pronóstico.

El objetivo de esta sección es analizar cómo cambian las métricas de la clase `No sobrevive 5 años` al modificar el umbral de decisión del modelo XGBoost optimizado.

### 7.8.1 Evaluación de distintos umbrales

El modelo XGBoost optimizado genera una probabilidad asociada a la clase positiva, es decir, a la clase `Sobrevive 5 años`.

Para obtener predicciones binarias, se aplica un umbral sobre dicha probabilidad:

- Si la probabilidad de supervivencia es mayor o igual al umbral, se predice `1`.
- Si la probabilidad de supervivencia es menor que el umbral, se predice `0`.

Reducir el umbral hace que más pacientes sean clasificados como supervivientes. Aumentarlo hace que el modelo sea más exigente para predecir supervivencia y, por tanto, clasifique más pacientes como no supervivientes.

Para seleccionar el umbral sin contaminar la evaluación final, el barrido de umbrales no se realiza sobre el conjunto de prueba, sino sobre el conjunto de entrenamiento mediante validación cruzada. Concretamente, se obtienen probabilidades *out-of-fold* (3 folds) sobre los datos de entrenamiento y se calculan las métricas de cada clase para cada umbral. De esta forma, el conjunto de prueba permanece intacto y no interviene en ningún momento en la elección del punto de corte.

```python
from sklearn.metrics import precision_recall_fscore_support
from sklearn.model_selection import cross_val_predict
from sklearn.base import clone

# Probabilidades out-of-fold sobre ENTRENAMIENTO.
# El umbral se selecciona sin usar en ningún momento el conjunto de test.
oof_proba_train = cross_val_predict(
    clone(best_xgb_model),
    X_train,
    y_train,
    cv=3,
    method="predict_proba",
    n_jobs=-1
)[:, 1]

threshold_results = []
thresholds = np.arange(0.30, 0.71, 0.02)

for threshold in thresholds:
    y_pred_threshold = (oof_proba_train >= threshold).astype(int)

    precision, recall, f1, support = precision_recall_fscore_support(
        y_train,
        y_pred_threshold,
        labels=[0, 1],
        zero_division=0
    )

    threshold_results.append({
        "threshold": round(threshold, 2),
        "precision_class_0": precision[0],
        "recall_class_0": recall[0],
        "f1_class_0": f1[0],
        "precision_class_1": precision[1],
        "recall_class_1": recall[1],
        "f1_class_1": f1[1],
        "accuracy": accuracy_score(y_train, y_pred_threshold)
    })

threshold_results = pd.DataFrame(threshold_results).round(4)
threshold_results
```

La tabla anterior, calculada sobre el conjunto de entrenamiento mediante validación cruzada, permite observar el compromiso entre precisión y recall al modificar el umbral de decisión. En particular, se presta atención al recall y al F1-score de la clase `0`, ya que esta clase representa a los pacientes que no alcanzan la supervivencia a cinco años.

```python
plt.figure(figsize=(9, 5))

plt.plot(
    threshold_results["threshold"],
    threshold_results["precision_class_0"],
    marker="o",
    label="Precision clase 0"
)

plt.plot(
    threshold_results["threshold"],
    threshold_results["recall_class_0"],
    marker="o",
    label="Recall clase 0"
)

plt.plot(
    threshold_results["threshold"],
    threshold_results["f1_class_0"],
    marker="o",
    label="F1 clase 0"
)

plt.title("Métricas de la clase No sobrevive 5 años según umbral")
plt.xlabel("Umbral de decisión para clase Sobrevive 5 años")
plt.ylabel("Valor de la métrica")
plt.ylim(0, 1)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

Una estrategia posible es seleccionar el umbral que maximiza el F1-score de la clase `No sobrevive 5 años`. Esta métrica equilibra la capacidad de detectar pacientes de alto riesgo con la precisión de dichas predicciones.

La selección se realiza utilizando las métricas obtenidas por validación cruzada sobre el conjunto de entrenamiento, de modo que el umbral queda fijado antes de evaluar el modelo sobre el conjunto de prueba.

Este criterio no pretende sustituir a la valoración clínica, pero permite adaptar la decisión del modelo al objetivo principal del proyecto: identificar adecuadamente a los pacientes con peor pronóstico.

Reducir el umbral respecto al valor por defecto aumenta la precisión de las predicciones de no supervivencia y mejora ligeramente su F1-score, pero reduce la proporción total de pacientes no supervivientes correctamente detectados.

### 7.8.3 Comparación entre umbral por defecto y umbral ajustado

Finalmente, se comparan las métricas principales del XGBoost optimizado con umbral por defecto y con el umbral ajustado.

```python
def metrics_for_threshold(
    y_true,
    y_pred,
    y_proba,
    model_name
):
    precision, recall, f1, _ = precision_recall_fscore_support(
        y_true,
        y_pred,
        labels=[0, 1],
        zero_division=0
    )

    return {
        "model": model_name,
        "precision_class_0": precision[0],
        "recall_class_0": recall[0],
        "f1_class_0": f1[0],
        "accuracy": accuracy_score(y_true, y_pred),
        "roc_auc": roc_auc_score(y_true, y_proba)
    }


threshold_comparison = pd.DataFrame([
    metrics_for_threshold(
        y_test,
        y_pred_best_xgb,
        y_proba_best_xgb,
        "XGBoost Optimizado - umbral 0.50"
    ),
    metrics_for_threshold(
        y_test,
        y_pred_best_xgb_threshold,
        y_proba_best_xgb,
        f"XGBoost Optimizado - umbral {best_threshold:.2f}"
    )
]).round(4)

threshold_comparison
```

La comparación entre umbrales muestra que el punto de corte modifica de forma relevante el comportamiento del modelo.

Con el umbral por defecto de 0,50, el XGBoost optimizado obtiene para la clase 0 una precisión de 0,4499, una sensibilidad de 0,7126 y un F1-score de 0,5516. La exactitud global es 0,7752.

Con el umbral ajustado de 0,38, la precisión de la clase 0 aumenta hasta 0,5466 y su F1-score hasta 0,5708. La exactitud global aumenta hasta 0,8257. Sin embargo, la sensibilidad de la clase 0 disminuye hasta 0,5972.

El AUC-ROC permanece constante en 0,8316, ya que depende de la ordenación de las puntuaciones y no del umbral utilizado para convertirlas en clases.

Por tanto, el umbral de 0,50 favorece una mayor detección de pacientes que no sobreviven cinco años, mientras que el umbral de 0,38 proporciona un equilibrio más favorable entre precisión y sensibilidad. Una elección aplicada requeriría definir previamente el coste relativo de cada tipo de error.

### 7.8.4 Intervalos de confianza de las métricas del modelo final

Para cuantificar la incertidumbre de las métricas calculadas con el umbral seleccionado, se aplica bootstrap sobre el conjunto de prueba. En cada remuestreo se calculan la exactitud, la precisión, la sensibilidad y el F1-score de la clase de no supervivencia, además del AUC-ROC.

```python
def bootstrap_final_metrics(
    y_true,
    y_score,
    threshold,
    n_boot=1000,
    seed=42
):
    y_true = np.asarray(y_true)
    y_score = np.asarray(y_score)

    y_pred = (y_score >= threshold).astype(int)

    point_estimates = {
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precisión clase 0": precision_score(
            y_true,
            y_pred,
            pos_label=0,
            zero_division=0
        ),
        "Sensibilidad clase 0": recall_score(
            y_true,
            y_pred,
            pos_label=0,
            zero_division=0
        ),
        "F1 clase 0": f1_score(
            y_true,
            y_pred,
            pos_label=0,
            zero_division=0
        ),
        "AUC-ROC": roc_auc_score(y_true, y_score)
    }

    bootstrap_values = {
        metric: []
        for metric in point_estimates
    }

    rng = np.random.default_rng(seed)
    n = len(y_true)

    for _ in range(n_boot):
        indices = rng.integers(0, n, size=n)

        y_sample = y_true[indices]
        score_sample = y_score[indices]
        pred_sample = (
            score_sample >= threshold
        ).astype(int)

        if np.unique(y_sample).size < 2:
            continue

        bootstrap_values["Accuracy"].append(
            accuracy_score(y_sample, pred_sample)
        )

        bootstrap_values["Precisión clase 0"].append(
            precision_score(
                y_sample,
                pred_sample,
                pos_label=0,
                zero_division=0
            )
        )

        bootstrap_values["Sensibilidad clase 0"].append(
            recall_score(
                y_sample,
                pred_sample,
                pos_label=0,
                zero_division=0
            )
        )

        bootstrap_values["F1 clase 0"].append(
            f1_score(
                y_sample,
                pred_sample,
                pos_label=0,
                zero_division=0
            )
        )

        bootstrap_values["AUC-ROC"].append(
            roc_auc_score(
                y_sample,
                score_sample
            )
        )

    rows = []

    for metric, estimate in point_estimates.items():
        lower, upper = np.percentile(
            bootstrap_values[metric],
            [2.5, 97.5]
        )

        rows.append({
            "metrica": metric,
            "estimacion": estimate,
            "IC95_inf": lower,
            "IC95_sup": upper
        })

    return pd.DataFrame(rows).round(4)


bootstrap_final_summary = bootstrap_final_metrics(
    y_true=y_test,
    y_score=y_proba_best_xgb,
    threshold=best_threshold,
    n_boot=1000,
    seed=42
)

bootstrap_final_summary
```

Estos intervalos cuantifican la variabilidad muestral interna de las métricas sobre la cohorte de prueba. No representan incertidumbre frente a cambios temporales, diferencias entre centros o poblaciones externas.

## 7.9 Análisis complementario de estabilidad mediante validación cruzada

La evaluación final se mantiene sobre la división estratificada entrenamiento-prueba (80/20), de modo que el conjunto de prueba permanece reservado para la estimación final del rendimiento. Como análisis complementario, se aplica una validación cruzada estratificada de 5 particiones (*5-fold*) **exclusivamente sobre el conjunto de entrenamiento**.

Para cada modelo se calcula el AUC-ROC en cada partición y se resume mediante la media, la desviación estándar y un intervalo aproximado del 95 % basado en la distribución *t*. Este análisis permite valorar la estabilidad interna ante distintas particiones aleatorias sin reutilizar el conjunto de prueba.

Los hiperparámetros del XGBoost ya fueron seleccionados previamente mediante validación cruzada de 3 particiones sobre el entrenamiento. Por ello, este bloque debe interpretarse como un análisis de estabilidad de la configuración elegida y no como una validación anidada completamente independiente. Una estimación estrictamente no sesgada de todo el proceso de selección requeriría *nested cross-validation*, cuyo coste computacional sería considerable para esta cohorte.

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.base import clone
from scipy.stats import t

cv_models = {
    "Regresión logística": clone(baseline_model),
    "Random Forest": clone(random_forest_model),
    "XGBoost optimizado": clone(best_xgb_model)
}

skf = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

cv_rows = []

for model_name, model in cv_models.items():
    scores = cross_val_score(
        model,
        X_train,
        y_train,
        cv=skf,
        scoring="roc_auc",
        n_jobs=1
    )

    mean_auc = scores.mean()
    std_auc = scores.std(ddof=1)

    t_critical = t.ppf(
        0.975,
        df=len(scores) - 1
    )

    margin = (
        t_critical
        * std_auc
        / np.sqrt(len(scores))
    )

    cv_rows.append({
        "modelo": model_name,
        "AUC_medio": mean_auc,
        "desv_std": std_auc,
        "IC95_aprox_inf": mean_auc - margin,
        "IC95_aprox_sup": mean_auc + margin,
        "AUC_min": scores.min(),
        "AUC_max": scores.max()
    })

cv_df = pd.DataFrame(cv_rows).round(4)
cv_df
```

La validación cruzada de cinco particiones confirma que la jerarquía observada en el conjunto de prueba se mantiene al variar las particiones del entrenamiento.

Los resultados fueron:

- Regresión logística: AUC medio de 0,8274, con desviación estándar de 0,0014 e intervalo aproximado de 0,8257–0,8291.
- Random Forest: AUC medio de 0,8214, con desviación estándar de 0,0016 e intervalo aproximado de 0,8194–0,8233.
- XGBoost optimizado: AUC medio de 0,8315, con desviación estándar de 0,0011 e intervalo aproximado de 0,8302–0,8329.

Las desviaciones reducidas indican estabilidad interna ante diferentes particiones aleatorias de la misma cohorte. No obstante, esta evidencia no sustituye una validación temporal o externa.

Los intervalos calculados a partir de cinco folds son aproximados y los folds no constituyen observaciones completamente independientes. Por ello, los intervalos bootstrap sobre el conjunto de prueba se mantienen como la principal cuantificación de incertidumbre del rendimiento final.

## 7.10 Evaluación y recalibración de las probabilidades

El AUC-ROC evalúa la capacidad de ordenar a los pacientes según su riesgo, pero no determina si las probabilidades estimadas coinciden con las frecuencias observadas. En un contexto clínico, discriminación y calibración son propiedades diferentes y ambas deben evaluarse.

Por ello se analiza el XGBoost optimizado mediante:

- *Brier score*, donde valores menores indican menor error cuadrático de probabilidad.
- Curva de calibración por deciles de riesgo.
- Tabla de calibración por deciles.
- Dos alternativas predefinidas de recalibración: regresión isotónica y escalado de Platt (*sigmoid*).

El XGBoost se entrena con `scale_pos_weight` para compensar el desbalance de clases. Esta ponderación puede mejorar la detección de la clase minoritaria, pero también desplazar las probabilidades sin calibrar. Los calibradores se ajustan exclusivamente con los datos de entrenamiento mediante validación cruzada interna (`cv=3`) y se evalúan una sola vez sobre el conjunto de prueba.

```python
from sklearn.calibration import CalibratedClassifierCV, calibration_curve
from sklearn.metrics import brier_score_loss, roc_auc_score

# Brier score del modelo sin calibrar (probabilidades ya calculadas sobre el test)
brier_uncal = brier_score_loss(y_test, y_proba_best_xgb)

# Recalibración isotónica y de Platt mediante validación cruzada interna (cv=3)
cal_isotonic = CalibratedClassifierCV(estimator=best_xgb_model, method="isotonic", cv=3)
cal_isotonic.fit(X_train, y_train)
proba_isotonic = cal_isotonic.predict_proba(X_test)[:, 1]

cal_platt = CalibratedClassifierCV(estimator=best_xgb_model, method="sigmoid", cv=3)
cal_platt.fit(X_train, y_train)
proba_platt = cal_platt.predict_proba(X_test)[:, 1]

calibration_summary = pd.DataFrame([
    {"metodo": "Sin calibrar", "brier_score": round(brier_uncal, 4),
     "auc": round(roc_auc_score(y_test, y_proba_best_xgb), 4)},
    {"metodo": "Isotónica", "brier_score": round(brier_score_loss(y_test, proba_isotonic), 4),
     "auc": round(roc_auc_score(y_test, proba_isotonic), 4)},
    {"metodo": "Platt (sigmoide)", "brier_score": round(brier_score_loss(y_test, proba_platt), 4),
     "auc": round(roc_auc_score(y_test, proba_platt), 4)},
])
calibration_summary
```

```python
plt.figure(figsize=(6.4, 6.0))
plt.plot([0, 1], [0, 1], "--", color="gray", lw=1, label="Calibración perfecta")

for proba, lab, col, mk in [
    (y_proba_best_xgb, "Sin calibrar", "#c0392b", "o"),
    (proba_isotonic, "Isotónica", "#2471a3", "s"),
    (proba_platt, "Platt (sigmoide)", "#2e7d4f", "^")
]:
    # strategy="quantile" construye grupos con tamaños similares (deciles de riesgo).
    frac, mean = calibration_curve(
        y_test,
        proba,
        n_bins=10,
        strategy="quantile"
    )
    brier = brier_score_loss(y_test, proba)
    plt.plot(
        mean,
        frac,
        marker=mk,
        color=col,
        lw=1.6,
        markersize=6,
        label=f"{lab} (Brier={brier:.3f})"
    )

plt.xlabel("Probabilidad media predicha (supervivencia a 5 años)")
plt.ylabel("Fracción observada de supervivientes")
plt.title("Curva de calibración por deciles — XGBoost optimizado")
plt.legend(loc="upper left")
plt.gca().set_aspect("equal")
plt.tight_layout()
plt.savefig("figuras/calibracion.png", dpi=300, bbox_inches="tight")
plt.show()
```

### 7.10.1 Calibración por deciles de riesgo

Para complementar la representación gráfica, se agrupan las predicciones en diez grupos de tamaño similar. En cada decil se comparan la probabilidad media estimada y la proporción observada de supervivientes. Cuanto menor sea la diferencia entre ambas cantidades, mejor será la calibración dentro de ese rango de riesgo.

```python
def calibration_by_decile(y_true, probabilities, method_name, n_bins=10):
    calibration_df = pd.DataFrame({
        "y_true": np.asarray(y_true),
        "probability": np.asarray(probabilities)
    })

    calibration_df["decil"] = pd.qcut(
        calibration_df["probability"],
        q=n_bins,
        labels=False,
        duplicates="drop"
    ) + 1

    summary = (
        calibration_df
        .groupby("decil", as_index=False)
        .agg(
            n=("y_true", "size"),
            probabilidad_media=("probability", "mean"),
            frecuencia_observada=("y_true", "mean")
        )
    )
    summary["diferencia_calibracion"] = (
        summary["probabilidad_media"] - summary["frecuencia_observada"]
    )
    summary.insert(0, "metodo", method_name)
    return summary

calibration_deciles = pd.concat([
    calibration_by_decile(y_test, y_proba_best_xgb, "Sin calibrar"),
    calibration_by_decile(y_test, proba_isotonic, "Isotónica"),
    calibration_by_decile(y_test, proba_platt, "Platt (sigmoide)")
], ignore_index=True)

calibration_deciles.round(4)
```

El modelo sin calibrar presenta un *Brier score* de 0,1604, mientras que la regresión isotónica y el escalado de Platt reducen este valor a 0,1105 y 0,1117, respectivamente. El AUC-ROC permanece prácticamente invariante, como es esperable: la recalibración modifica la correspondencia entre puntuaciones y frecuencias observadas, pero apenas altera el orden de los pacientes.

La curva y la tabla por deciles permiten identificar en qué zonas de riesgo se concentran las desviaciones. Los resultados indican que las puntuaciones originales del modelo ponderado no deben interpretarse directamente como probabilidades. Las dos técnicas de recalibración mejoran de forma sustancial la calibración interna sobre la cohorte estudiada.

No se selecciona una de las dos técnicas como calibrador clínico definitivo basándose únicamente en su rendimiento sobre el conjunto de prueba. Para elegir y desplegar un método de calibración sería necesario hacerlo mediante validación interna específica y, posteriormente, comprobar su estabilidad en una cohorte temporal o externa. En consecuencia, las probabilidades recalibradas deben describirse como **mejor calibradas internamente**, no como probabilidades clínicas generalizables.

## 7.11 Análisis de sensibilidad sobre dimensionalidad y valores no informados

Para comprobar si las decisiones relacionadas con la alta cardinalidad y la ausencia de información condicionan el rendimiento, se construye una versión alternativa del conjunto de predictores.

En este análisis:

- Se excluye `tumor_size`, debido a su elevada proporción de valores no informados.
- Se agrupa el estado civil en categorías más amplias.
- Se conservan los grupos de edad habituales, pero las categorías con menos de 500 observaciones de entrenamiento se agrupan automáticamente.
- Las categorías infrecuentes de `histologic_type` y `surgery_primary_site` también se combinan mediante `OneHotEncoder`.
- Se utiliza exactamente la misma división entrenamiento-prueba que en el análisis principal.
- Se mantienen los mismos algoritmos e hiperparámetros.

Este bloque constituye un análisis de sensibilidad y no un nuevo proceso de selección del modelo basado en el conjunto de prueba.

```python
from sklearn.base import clone

# Conjunto alternativo de predictores
X_sensitivity = df_model_main.drop(
    columns=[
        "survived_5_years",
        "tumor_size"
    ]
).copy()

# Agrupación del estado civil
marital_mapping = {
    "Married (including common law)": "Married",
    "Single (never married)": "Single/Unmarried",
    "Unmarried or Domestic Partner": "Single/Unmarried",
    "Divorced": "Divorced/Separated",
    "Separated": "Divorced/Separated",
    "Widowed": "Widowed",
    "Unknown": "Unknown/Other"
}

X_sensitivity["marital_status"] = (
    X_sensitivity["marital_status"]
    .replace(marital_mapping)
)

categorical_features_sensitivity = [
    "age_group",
    "sex",
    "race",
    "marital_status",
    "primary_site",
    "histologic_type",
    "summary_stage",
    "surgery_primary_site",
    "radiation",
    "chemotherapy"
]

numeric_features_sensitivity = [
    "year_diagnosis"
]

# Se reutilizan exactamente los índices de train y test del análisis principal
X_train_sensitivity = X_sensitivity.loc[
    X_train.index
].copy()

X_test_sensitivity = X_sensitivity.loc[
    X_test.index
].copy()

# Conversión homogénea a texto de los predictores categóricos
X_train_sensitivity[
    categorical_features_sensitivity
] = X_train_sensitivity[
    categorical_features_sensitivity
].astype(str)

X_test_sensitivity[
    categorical_features_sensitivity
] = X_test_sensitivity[
    categorical_features_sensitivity
].astype(str)

categorical_transformer_sensitivity = Pipeline(steps=[
    (
        "imputer",
        SimpleImputer(strategy="most_frequent")
    ),
    (
        "encoder",
        OneHotEncoder(
            handle_unknown="infrequent_if_exist",
            min_frequency=500
        )
    )
])

numeric_transformer_sensitivity = Pipeline(steps=[
    (
        "scaler",
        StandardScaler()
    )
])

preprocessor_sensitivity = ColumnTransformer(
    transformers=[
        (
            "categorical",
            categorical_transformer_sensitivity,
            categorical_features_sensitivity
        ),
        (
            "numeric",
            numeric_transformer_sensitivity,
            numeric_features_sensitivity
        )
    ]
)

logistic_sensitivity = Pipeline(steps=[
    (
        "preprocessor",
        clone(preprocessor_sensitivity)
    ),
    (
        "classifier",
        clone(
            baseline_model.named_steps["classifier"]
        )
    )
])

xgb_sensitivity = Pipeline(steps=[
    (
        "preprocessor",
        clone(preprocessor_sensitivity)
    ),
    (
        "classifier",
        clone(
            best_xgb_model.named_steps["classifier"]
        )
    )
])

logistic_sensitivity.fit(
    X_train_sensitivity,
    y_train
)

xgb_sensitivity.fit(
    X_train_sensitivity,
    y_train
)

proba_logistic_sensitivity = (
    logistic_sensitivity
    .predict_proba(X_test_sensitivity)[:, 1]
)

proba_xgb_sensitivity = (
    xgb_sensitivity
    .predict_proba(X_test_sensitivity)[:, 1]
)

pred_logistic_sensitivity = (
    logistic_sensitivity
    .predict(X_test_sensitivity)
)

pred_xgb_sensitivity = (
    xgb_sensitivity
    .predict(X_test_sensitivity)
)

n_features_main = len(
    baseline_model
    .named_steps["preprocessor"]
    .get_feature_names_out()
)

n_features_sensitivity = len(
    logistic_sensitivity
    .named_steps["preprocessor"]
    .get_feature_names_out()
)

sensitivity_results = pd.DataFrame([
    {
        "modelo": "Regresión logística principal",
        "n_predictores_originales": X_train.shape[1],
        "n_caracteristicas_transformadas": n_features_main,
        "AUC_ROC": roc_auc_score(
            y_test,
            y_proba_baseline
        ),
        "accuracy": accuracy_score(
            y_test,
            y_pred_baseline
        ),
        "F1_clase_0": f1_score(
            y_test,
            y_pred_baseline,
            pos_label=0
        )
    },
    {
        "modelo": "Regresión logística sensibilidad",
        "n_predictores_originales": X_train_sensitivity.shape[1],
        "n_caracteristicas_transformadas": n_features_sensitivity,
        "AUC_ROC": roc_auc_score(
            y_test,
            proba_logistic_sensitivity
        ),
        "accuracy": accuracy_score(
            y_test,
            pred_logistic_sensitivity
        ),
        "F1_clase_0": f1_score(
            y_test,
            pred_logistic_sensitivity,
            pos_label=0
        )
    },
    {
        "modelo": "XGBoost optimizado principal",
        "n_predictores_originales": X_train.shape[1],
        "n_caracteristicas_transformadas": n_features_main,
        "AUC_ROC": roc_auc_score(
            y_test,
            y_proba_best_xgb
        ),
        "accuracy": accuracy_score(
            y_test,
            y_pred_best_xgb
        ),
        "F1_clase_0": f1_score(
            y_test,
            y_pred_best_xgb,
            pos_label=0
        )
    },
    {
        "modelo": "XGBoost optimizado sensibilidad",
        "n_predictores_originales": X_train_sensitivity.shape[1],
        "n_caracteristicas_transformadas": n_features_sensitivity,
        "AUC_ROC": roc_auc_score(
            y_test,
            proba_xgb_sensitivity
        ),
        "accuracy": accuracy_score(
            y_test,
            pred_xgb_sensitivity
        ),
        "F1_clase_0": f1_score(
            y_test,
            pred_xgb_sensitivity,
            pos_label=0
        )
    }
]).round(4)

sensitivity_results
```

El análisis de sensibilidad reduce el espacio de características transformadas de 487 a 113 mediante la exclusión de `tumor_size`, la agrupación del estado civil y la combinación automática de categorías infrecuentes.

En la regresión logística, el AUC-ROC pasa de 0,8274 a 0,8267, una diferencia de solo 0,0007 puntos. En el XGBoost optimizado, el AUC pasa de 0,8316 a 0,8315, una diferencia de 0,0001 puntos.

La reducción de dimensionalidad produce, por tanto, cambios mínimos en la capacidad discriminativa. Esto indica que el rendimiento no depende de forma relevante de mantener `tumor_size`, pese a su elevada proporción de información ausente, ni de conservar todas las categorías infrecuentes por separado.

Estos resultados apoyan la robustez general de las conclusiones frente a decisiones alternativas de preprocesamiento. No obstante, el análisis se interpreta como una comprobación de sensibilidad y no como una nueva selección retrospectiva del modelo a partir del conjunto de prueba.

# 8. Interpretabilidad del modelo

En problemas aplicados al ámbito sanitario, la evaluación de un modelo predictivo no debe limitarse al análisis de métricas de rendimiento. También es necesario comprender qué variables influyen en las predicciones y si los patrones aprendidos son coherentes desde el punto de vista clínico.

La interpretabilidad resulta especialmente importante en este trabajo por tres motivos:

- El problema se centra en la predicción de supervivencia a cinco años en cáncer de mama, un desenlace clínicamente relevante.
- El modelo podría utilizarse como herramienta de apoyo a la estratificación de riesgo, por lo que sus predicciones deben ser justificables.
- Algunas variables del dataset representan códigos clínicos, valores especiales o tratamientos, cuya interpretación requiere cautela.

En esta sección se analizan dos modelos:

- La regresión logística baseline, por su simplicidad e interpretabilidad directa mediante coeficientes.
- El XGBoost optimizado, por ser el modelo con mayor AUC-ROC y mejor capacidad discriminativa global.

## 8.1 Interpretabilidad de la regresión logística

La regresión logística permite analizar la contribución de cada variable transformada mediante sus coeficientes.

En este modelo, la clase positiva corresponde a `survived_5_years = 1`, es decir, pacientes que sobreviven al menos cinco años. Por tanto:

- Coeficientes positivos se asocian con mayor probabilidad de supervivencia a cinco años.
- Coeficientes negativos se asocian con menor probabilidad de supervivencia a cinco años.

Dado que las variables categóricas han sido codificadas mediante One-Hot Encoding, cada coeficiente corresponde a una categoría concreta de una variable original.

```python
logistic_preprocessor = baseline_model.named_steps["preprocessor"]
logistic_classifier = baseline_model.named_steps["classifier"]

logistic_feature_names = logistic_preprocessor.get_feature_names_out()
logistic_coefficients = logistic_classifier.coef_[0]

coef_df = pd.DataFrame({
    "feature": logistic_feature_names,
    "coefficient": logistic_coefficients
})

coef_df["abs_coefficient"] = coef_df["coefficient"].abs()

coef_df = coef_df.sort_values(
    by="abs_coefficient",
    ascending=False
)

coef_df.head(20)
```

La tabla anterior muestra las variables transformadas con mayor peso absoluto en el modelo de regresión logística. Los coeficientes con mayor magnitud son los que más influyen en la predicción del modelo, aunque su interpretación debe realizarse teniendo en cuenta que las variables categóricas han sido codificadas mediante One-Hot Encoding.

Esta formulación fue una decisión metodológica consciente. Permitió comparar modelos de clasificación habituales mediante un desenlace fijo, operativo y fácilmente interpretable, así como analizar el desbalance, ajustar el umbral de decisión y estudiar los tipos de error dentro de un marco común.

La definición requirió tratar adecuadamente la censura. Los pacientes vivos con menos de 60 meses de seguimiento fueron excluidos porque no era posible determinar de forma fiable si alcanzarían el horizonte de cinco años. La contrapartida es la pérdida de información sobre el tiempo exacto hasta el evento y la exclusión de observaciones censuradas, información que podría aprovecharse mediante modelos específicos de supervivencia.

Tras restringir el dataset al periodo 2000--2017, la variable objetivo presentó una distribución moderadamente desbalanceada:

- 803.498 pacientes sobrevivieron al menos cinco años, representando el 80,60 %.
- 193.418 pacientes fallecieron antes de cinco años, representando el 19,40 %.

Este desbalance condicionó la selección de métricas y justificó prestar especial atención a la clase minoritaria, correspondiente a pacientes con peor pronóstico.

Con el umbral por defecto de 0,50, la sensibilidad de la clase 0 era mayor, con un valor de 0,7126, pero su precisión era inferior, con 0,4499. El umbral de 0,38 mejora la precisión y el F1-score, a costa de dejar sin detectar una mayor proporción de pacientes que no sobreviven cinco años.

Por tanto, el umbral ajustado no constituye una mejora universal, sino un punto de operación diferente. Una selección aplicada debería basarse en una valoración explícita de las consecuencias relativas de los falsos positivos y falsos negativos.

## 10.6 Conclusiones sobre calibración

La evaluación de calibración mostró que el AUC-ROC no es suficiente para interpretar las puntuaciones del modelo como probabilidades. El XGBoost sin calibrar obtuvo un *Brier score* de 0,1604 y presentó desviaciones relevantes entre las probabilidades estimadas y las frecuencias observadas.

Por tanto, las puntuaciones directas del XGBoost ponderado no deben interpretarse como probabilidades clínicas. Las versiones recalibradas presentan una calibración interna sustancialmente mejor, pero su uso clínico requeriría validar tanto el modelo como el calibrador en periodos posteriores o en una población externa.

## 10.7 Conclusiones sobre interpretabilidad

El análisis de interpretabilidad permitió comprobar que los modelos se apoyan en variables clínicamente plausibles.

En la regresión logística, los coeficientes positivos se asociaron con mayor probabilidad estimada de supervivencia a cinco años. Entre ellos destacaron la enfermedad localizada, edades intermedias, tamaños tumorales pequeños y determinados códigos quirúrgicos e histológicos.

Los coeficientes negativos se asociaron con menor probabilidad estimada de supervivencia. Entre los factores más relevantes aparecieron la edad avanzada, la enfermedad distante, determinados códigos histológicos, códigos quirúrgicos asociados a peor pronóstico y el rechazo de radioterapia.

En el XGBoost optimizado, las variables más importantes estuvieron relacionadas con cirugía, estadio, edad, estado civil, histología, radioterapia y raza. Destacaron especialmente `surgery_primary_site_0`, `summary_stage_Distant` y grupos de edad avanzada.

No obstante, algunas variables importantes requieren interpretación cautelosa, especialmente aquellas relacionadas con códigos administrativos, valores desconocidos o tratamientos observacionales.

## 10.8 Conclusiones del análisis de errores

El análisis de errores permitió estudiar en mayor detalle los casos en los que el modelo falla.

El modelo XGBoost optimizado con umbral ajustado clasificó correctamente a la mayoría de pacientes supervivientes y mostró un equilibrio razonable entre precisión y recall para la clase de no supervivencia. Sin embargo, persistieron errores clínicamente relevantes.

Los falsos positivos, es decir, pacientes que realmente no sobreviven cinco años pero son clasificados como supervivientes, representan un tipo de error especialmente delicado, ya que implican no identificar adecuadamente a pacientes de peor pronóstico.

El análisis por probabilidades mostró que los verdaderos negativos suelen recibir probabilidades bajas de supervivencia, mientras que los verdaderos positivos presentan probabilidades elevadas. Sin embargo, algunos falsos positivos reciben probabilidades moderadas o altas de supervivencia, lo que sugiere la existencia de factores pronósticos no recogidos por las variables disponibles.

El análisis por estadio mostró que el modelo identifica bien el mal pronóstico de la enfermedad distante y el buen pronóstico de la enfermedad localizada, aunque presenta más dificultad en casos regionales, no estadificados o localizados con desenlace desfavorable.

Por edad, el modelo funciona mejor en grupos intermedios y tiende a asumir peor pronóstico en edades avanzadas. Esto genera errores en pacientes mayores que finalmente sí sobreviven cinco años.

En conjunto, el análisis de errores confirma que el modelo aprende patrones coherentes, pero también evidencia limitaciones propias de los datos disponibles.

## 10.9 Limitaciones del trabajo

Aunque los resultados obtenidos son metodológicamente coherentes, el trabajo presenta varias limitaciones.

En primer lugar, el estudio se basa en datos observacionales procedentes de un registro poblacional. Las asociaciones encontradas no deben interpretarse como relaciones causales. Las variables de tratamiento pueden estar condicionadas por el estadio, la edad, el estado general, las preferencias terapéuticas y otros factores no observados.

En segundo lugar, la clasificación binaria a cinco años simplifica la naturaleza temporal de la supervivencia. Esta elección facilita un problema supervisado directo, pero pierde el tiempo exacto hasta el evento y obliga a excluir pacientes censurados antes de 60 meses. Modelos como Cox, Random Survival Forest, XGBoost survival, DeepSurv o enfoques de riesgos competitivos permitirían aprovechar esa información.

En tercer lugar, algunas variables presentan una proporción elevada de valores desconocidos o no informados. Destaca `tumor_size`, cuya categoría `Blank(s)` limita la interpretación clínica de su contribución. Además, el dataset no incluye biomarcadores moleculares, comorbilidades, estado funcional, recurrencias, adherencia, información genética ni detalles terapéuticos completos.

En cuarto lugar, la evaluación principal utiliza una partición entrenamiento-prueba interna, complementada con validación cruzada estratificada de cinco particiones sobre el conjunto de entrenamiento. Ambas evaluaciones proceden de la misma cohorte y del mismo periodo temporal. Por tanto, muestran capacidad de generalización interna, pero no demuestran estabilidad en años posteriores ni en otras poblaciones. También debe considerarse que `year_diagnosis` puede capturar tendencias temporales, cambios terapéuticos o modificaciones del registro.

En quinto lugar, la optimización del XGBoost y el análisis posterior de estabilidad no constituyen una validación cruzada anidada completa. Una *nested cross-validation* reduciría el sesgo asociado a la selección de hiperparámetros, aunque tendría un coste computacional elevado para una muestra cercana al millón de observaciones.

En sexto lugar, la calibración se evaluó de forma interna. La regresión isotónica y Platt mejoraron el *Brier score* sobre el conjunto de prueba, pero no se ha demostrado que esa calibración se conserve en periodos posteriores o en una cohorte externa. Por ello, las probabilidades recalibradas no deben considerarse todavía probabilidades clínicas generalizables.

Finalmente, los modelos desarrollados deben entenderse como herramientas experimentales de apoyo a la estratificación de riesgo, no como sistemas autónomos de decisión clínica.

## 10.10 Conclusión general

El notebook ha desarrollado un *pipeline* completo para estimar la supervivencia a cinco años en pacientes con cáncer de mama mediante datos clínico-epidemiológicos del programa SEER.

El proceso incluye carga y comprensión del dataset, limpieza, tratamiento de valores especiales, definición rigurosa de la variable objetivo, análisis exploratorio, preprocesamiento, entrenamiento y optimización de modelos, ajuste de umbral, análisis complementario mediante validación cruzada, evaluación de calibración, interpretabilidad y análisis de errores.

La formulación como clasificación binaria se presenta como una decisión metodológica delimitada: permite comparar modelos dentro de un marco operativo común, aunque no sustituye a un análisis de supervivencia y requiere excluir observaciones censuradas. El XGBoost optimizado alcanzó el mayor AUC-ROC, mientras que la regresión logística mantuvo un rendimiento muy próximo con mayor simplicidad e interpretabilidad.

La evaluación muestra también que discriminación y calibración deben analizarse por separado. Las puntuaciones originales del XGBoost ponderado no estaban adecuadamente calibradas, pero la regresión isotónica y el escalado de Platt mejoraron sustancialmente la correspondencia interna entre probabilidades estimadas y frecuencias observadas.

El análisis de interpretabilidad y de errores confirma que los modelos aprenden patrones coherentes con el conocimiento clínico, especialmente relacionados con edad, estadio, cirugía e histología. Al mismo tiempo, la ausencia de validación temporal y externa, la naturaleza observacional de las variables y la simplificación del desenlace limitan cualquier traslación clínica directa.

En conjunto, el trabajo proporciona una base sólida, reproducible y metodológicamente prudente para la predicción a horizonte fijo. La extensión natural consiste en comparar este enfoque con modelos específicos de supervivencia y validar temporal o externamente tanto el rendimiento como la calibración.
